```python
from databricks.vector_search.client import VectorSearchClient

client = VectorSearchClient()
table_name = "yyang.hackathon_2025q3_project_geniepromptautocompletion_genieeng_feindustryhls.questions_table"
index_name = table_name.replace("questions_table", "questions_table_vs")

vs_endpoint_name = "vs_endpoint_2025q3_hackathon_project_geniepromptautocompletion_genieeng_feindustryhls".lower()[:49]

# Check if endpoint exists
endpoints = client.list_endpoints()['endpoints']
endpoint_names = [ep['name'] for ep in endpoints]

if vs_endpoint_name in endpoint_names:
    print(f"VS endpoint '{vs_endpoint_name}' already exists.")
else:
    client.create_endpoint(name=vs_endpoint_name, endpoint_type="STANDARD")
    print(f"Created VS endpoint '{vs_endpoint_name}'.")
```

```python
spark.sql(f"ALTER TABLE {table_name} SET TBLPROPERTIES (delta.enableChangeDataFeed = true)")
```

```python
client.wait_for_endpoint(vs_endpoint_name, "READY")

index = client.create_delta_sync_index(
    endpoint_name=vs_endpoint_name,
    source_table_name=table_name,
    index_name=index_name,
    pipeline_type="Continuous", # "TRIGGERED",
    primary_key="id",
    embedding_source_column="query",
    embedding_model_endpoint_name="databricks-gte-large-en"
)
```

```python
from databricks.vector_search.client import VectorSearchClient

vsc = VectorSearchClient()
```

```python
table_name = "yyang.hackathon_2025q3_project_geniepromptautocompletion_genieeng_feindustryhls.questions_table"

vs_index_fullname = table_name.replace("questions_table", "questions_table_vs")

vs_endpoint_name = "vs_endpoint_2025q3_hackathon_project_geniepromptautocompletion_genieeng_feindustryhls".lower()[:49]
```

```python
index = vsc.get_index(endpoint_name=vs_endpoint_name, index_name=vs_index_fullname)

index.describe()
```

```python
# Wait for index to come online. Expect this command to take several minutes.
import time
while not index.describe().get('status').get('detailed_state').startswith('ONLINE'):
  print("Waiting for index to be ONLINE...")
  time.sleep(5)
print("Index is ONLINE")
index.describe()
```

```python
results = index.similarity_search(
  query_text="What is the phone number of highest SAT Score",
  columns="query",
  num_results=5)

results
```
